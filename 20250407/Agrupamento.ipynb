```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#gerando dados sinteticos
X, y_true = make_blobs(
    n_samples=300,
    centers = 4,
    cluster_std=0.60,
    random_state=0
)

print(f"\n Fromado dos dados{X.shape}")
plt.scatter(X[:,0],X[:, 1], c=y_true, s=50,cmap ='viridis', alpha=0.7)
plt.title('Dados originais com clusters verdadeiros')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_true, s=50,cmap ='viridis', alpha=0.7)
centers = kmeans.cluster_centers_

plt.scatter(centers[:,0],centers[:,1], c='red', s=200, alpha=0.75, marker='X')
plt.title('Agrupamento K-means')
plt.show()
```

```python
print("\n Centroides dos clusters:")
for i, center in enumerate(centers):
  print(f"Cluster{i}: ({center[0]:.2f}, {center[1]:.2f})")
```
