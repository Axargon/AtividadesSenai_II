# Exemplo de uso de Regressão Linear

Método estatístico utilizado para modelar a relação linear entre uma variável dependente (ou resposta) e uma ou mais variáveis independentes (ou preditoras), ajustando uma equação da forma =0+β1x+ϵy=β 0​ +β 1​x+ϵ, onde y
y é a variável dependente, x é a variável independente, β0 ​e β1 são os coeficientes que representam o intercepto e a inclinação da reta, respectivamente, e ϵ é o erro aleatório. O objetivo é encontrar os coeficientes que minimizam a soma dos quadrados dos resíduos, proporcionando a melhor reta de ajuste para os dados observados.

```python
#Importar as bibliotecas apropriadas
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Criação de dados randomicos para simulação de um conjunto de dados que representam imoveis com numero de quartos e preco destes imoveis
np.random.seed(42)
X = 2 * np.random.rand(100,1) # numero de quartos que vai de 0 a 2
y = 4+3*2 *X + np.random.randn(100,1) #preco da casa com ruido

#Imprimir array de quartos e de preços (verificar coerencia)
print("\nPrimeiros 5 exemplos: ")
print(f"X (Quartos): {X[:5].flatten()}")
print(f"y (Quartos): {y[:5].flatten()}")

# Plotar o grafico de scatter
plt.scatter(X,y,alpha=0.7)
plt.xlabel('Número de quartos')
plt.ylabel('Preço cda casa')
plt.title('Relação de Quartos vs Preço')
plt.show()
```

Agora que temos os dados aleatórios, podemos construir um modelo de predição do tipo de regressão linear.

1.   Instanciamos o modelo contido na biblioteca importada
2.   Utilizamos os dados aleatorios que simulam o caso e treinamos o modelo


```python
#criacao e treinamento
model = LinearRegression()
model.fit(X,y)

#Verificação dos coeficientes linear e angular
print(f"\nIntercepto(b):{model.intercept_[0]:.2f}")
print(f"\nCoeficiente angular(a):{model.coef_[0][0]:.2f}")
```

```python
# Previsoes
X_new = np.array([[1.5]]) # uma casa com 1.5 quartos
y_pred = model.predict(X_new)
print(f"Previsao para casa com 1.5 quartos: {y_pred[0][0]:.2f} x 100k")


#Plot dos dados com o modelo de regressão treinado
plt.scatter(X,y,alpha=0.7)
plt.plot(X, model.predict(X), color ='red',linewidth=3)
plt.xlabel('Número de quartos')
plt.ylabel('Preço cda casa(x100k)')
plt.title('Regressão Linear')
plt.show()
```
